```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python

img = cv2.imread('messi.jpg')

plt.figure(figsize=[20, 10])
plt.subplot(131)
plt.imshow(img)
plt.title("bgr")
plt.subplot(132)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("rgb")
plt.subplot(133)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray,cmap='gray')
plt.title("gray")
```

# Drawing shapes:

```python
black_square = np.zeros((512,512,3), np.uint8)
cv2.line(black_square, (0,0), (511,511), (255,0,0),8)
cv2.rectangle(black_square, (350,2), (510,150), (0,255,0), 8)
cv2.circle(black_square ,(430,75), 75, (0,0,255), -1)
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_square,'OpenCV',(10,500), font, 4,(255,255,255),15,cv2.LINE_AA)
cv2.ellipse(black_square,(256,256),(100,50),320,0,180,255,-1)

pts = np.array([[1,5],[220,40],[300,150],[130,200]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(black_square,[pts],True,(0,255,255), 5)
plt.imshow(black_square)
```

# ROI selector

```python
x, y, w, h = cv2.selectROI(img, False)
cropped_img = img_rgb[y:y+h, x:x+w]
cv2.imshow("Image", cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pasted_ball = img_rgb
pasted_ball[y:y+h, 20:w+20] = cropped_img
plt.imshow(pasted_ball)
```
